```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
drive.mount('/content/drive')  # Mount Google Drive

# Load the dataset (update path if needed)
file_path = "/content/drive/My Drive/Road Accident Data.csv"
df = pd.read_csv(file_path)

# Show basic info
print(df.head())
print(df.info())
```

```python
# Total number of accidents
total_accidents = df.shape[0]
print(f"Total Accidents: {total_accidents}")

# Accidents by Year, Month, Day
df["Year"] = df["Year"].astype(int)  # Ensure Year is integer

plt.figure(figsize=(12,5))
sns.countplot(x="Year", data=df, palette="coolwarm")
plt.title("Accidents per Year")
plt.xlabel("Year")
plt.ylabel("Count of Accidents")
plt.show()

plt.figure(figsize=(12,5))
sns.countplot(x="Month", data=df, palette="viridis")
plt.title("Accidents per Month")
plt.xlabel("Month")
plt.ylabel("Count of Accidents")
plt.show()

plt.figure(figsize=(10,5))
sns.countplot(x="Day_of_Week", data=df, order=["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])
plt.title("Accidents per Day of the Week")
plt.xlabel("Day")
plt.ylabel("Count of Accidents")
plt.show()
```

```python
# Top 10 locations with highest accidents
df["Local_Authority_(District)"].value_counts().head(10).plot(kind="bar", figsize=(12,5), color="red")
plt.title("Top 10 Districts with Highest Accidents")
plt.xlabel("District")
plt.ylabel("Accident Count")
plt.show()
```

```python
# Severity distribution
plt.figure(figsize=(8,5))
sns.countplot(x="Accident_Severity", data=df, palette="magma")
plt.title("Accident Severity Distribution")
plt.xlabel("Severity")
plt.ylabel("Count")
plt.show()

# Percentage of fatal accidents
fatal_accidents = df[df["Accident_Severity"] == "Fatal"].shape[0]
fatal_percentage = (fatal_accidents / total_accidents) * 100
print(f"Fatal Accidents Percentage: {fatal_percentage:.2f}%")
```

```python
# Check age distribution
plt.figure(figsize=(10,5))
sns.histplot(df["Number_of_Casualties"], bins=20, kde=True, color="blue")
plt.title("Age Distribution of Casualties")
plt.xlabel("Age")
plt.ylabel("Number of Accidents")
plt.show()

# Gender distribution
df["Vehicle_Type"].value_counts().plot(kind="bar", color="green", figsize=(10,5))
plt.title("Accidents by Vehicle Type (Proxy for Gender)")
plt.xlabel("Vehicle Type")
plt.ylabel("Accident Count")
plt.show()
```

```python
# Accidents by weather conditions
plt.figure(figsize=(10,5))
sns.countplot(x="Weather_Conditions", data=df, palette="coolwarm")
plt.title("Accidents by Weather Conditions")
plt.xticks(rotation=45)
plt.show()

# Accidents by road type
plt.figure(figsize=(10,5))
sns.countplot(x="Road_Type", data=df, palette="viridis")
plt.title("Accidents by Road Type")
plt.xticks(rotation=45)
plt.show()
```

```python
# Accidents by Vehicle Type
plt.figure(figsize=(10,5))
sns.countplot(y="Vehicle_Type", data=df, palette="inferno", order=df["Vehicle_Type"].value_counts().index)
plt.title("Accidents by Vehicle Type")
plt.show()
```

```python
# Extract hour from Time column
df["Time"] = pd.to_datetime(df["Time"], errors="coerce")
df["Hour"] = df["Time"].dt.hour

# Accidents by Hour
plt.figure(figsize=(10,5))
sns.histplot(df["Hour"], bins=24, kde=True, color="purple")
plt.title("Accidents by Hour of the Day")
plt.xlabel("Hour")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
df["Carriageway_Hazards"].value_counts().plot(kind="bar", figsize=(12,5), color="red")
plt.title("Most Common Accident Causes")
plt.xlabel("Cause")
plt.ylabel("Accident Count")
plt.show()
```

```python
df["Number_of_Casualties"].value_counts().plot(kind="bar", figsize=(10,5), color="brown")
plt.title("Distribution of Casualties")
plt.xlabel("Number of Casualties")
plt.ylabel("Accident Count")
plt.show()
```

```python
sns.countplot(x="Urban_or_Rural_Area", data=df, palette="coolwarm")
plt.title("Urban vs Rural Accidents")
plt.show()
```
